## Ensuring all columns have the correct data types

```python
# Handling non-numeric values in numeric columns
# Targeting columns that are currently numeric (int or float)
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()

anomalies_found = 0

print("--- Checking and Fixing Non-Numeric Data in Numeric Columns ---")

for col in numeric_cols:
    # 1. Temporarily coercing the column to numeric, turning non-convertible strings into NaN
    temp_numeric = pd.to_numeric(df[col], errors='coerce')

    # 2. Identifying where conversion failed AND the original value was NOT already NaN
    # This isolates the actual 'wrong kind of data' (e.g., '---' or 'error code')
    is_anomaly = temp_numeric.isna() & df[col].notna()
    
    count_of_bad_data = is_anomaly.sum()
    
    if count_of_bad_data > 0:
        anomalies_found += count_of_bad_data
        
        # 3. Applying the fix: Overwrite the original column, replacing bad data with NaN
        df[col] = temp_numeric 
        
        print(f"| FIXED: '{col}' - {count_of_bad_data} non-numeric values converted to NaN.")

print("\n--- Summary (Part 1) ---")
if anomalies_found > 0:
    print(f"Successfully converted {anomalies_found} structural errors to NaN across the dataset.")
else:
    print("✅ No non-numeric data anomalies found in current numeric columns.")
```

```python
# Forcing all the elements in categorical columns into category, to ensure consistent behaviour later on
# Selects all columns that Pandas currently registers as 'object'
object_cols = df.select_dtypes(include=['object']).columns

print("--- Enforcing String Type for Object Columns ---")

for col in object_cols:
    # Convert every value in the column to a string
    df[col] = df[col].astype(str)
    
    if df[col].nunique() < 50: 
        df[col] = df[col].astype('category')
        print(f"| Column '{col}' converted to category.")
    else:
        print(f"| Column '{col}' converted to object (string).")

print("--- All target columns are now consistently string/category type. ---")
```

## Handling null values

. Handling null values in numerical columns

```python
# Handling null values in numerical columns
# 1. Identifying all numerical columns (int and float)
numerical_cols = df.select_dtypes(include=['number']).columns

print("--- Starting Numerical Imputation (Median) ---")
imputed_cols = 0

for col in numerical_cols:
    # 2. Checking if the column has any missing values
    if df[col].isnull().any():
        
        # 3. Calculating the median for the current column
        # .median() ignores NaN values automatically when calculating.
        median_value = df[col].median()
        
        # 4. Filling the NaN values with the calculated median
        df[col].fillna(median_value, inplace=True)
        
        # Counting the number of cells imputed
        num_imputed = df[col].isnull().sum()
        
        # Re-checking to confirm successful imputation (should be 0)
        if num_imputed == 0:
            print(f"| Imputed '{col}' with median value ({median_value:.2f}).")
            imputed_cols += 1
        else:
            print(f"| WARNING: Column '{col}' failed to fully impute.")
            
print("\n--- Summary ---")
if imputed_cols > 0:
    print(f"✅ Successfully imputed missing values in {imputed_cols} numerical columns.")
else:
    print("ℹ️ No missing values found in the numerical columns or all were already handled.")
```

.. Handling null values in categorical columns

```python
# Will replace all nan items with missing as that lowers the risk of bias and overfitting
```

```python
import pandas as pd

# Defining the string that represents the missing value in categorical columns
MISSING_VALUE_STRING = 'nan'

# Defining the new, informative category name
NEW_CATEGORY_NAME = 'Bilinmiyor'

# 1. Identifying all non-numerical columns
categorical_cols = df.select_dtypes(exclude=['number']).columns

print("--- Starting Categorical Imputation (FIXED) ---")
imputed_cols = 0

for col in categorical_cols:
    # Checking if the column is a true 'category' dtype
    is_categorical_dtype = df[col].dtype.name == 'category'
    
    # 2. Checking if the new category is already registered (prevents error if run twice)
    if is_categorical_dtype and NEW_CATEGORY_NAME not in df[col].cat.categories:
        # Adding the new category to the column's master list
        df[col] = df[col].cat.add_categories(NEW_CATEGORY_NAME)
        print(f"| Added '{NEW_CATEGORY_NAME}' to categories for column '{col}'.")
    
    # 3. Creating a Boolean mask to find where the value equals the missing string
    is_missing = (df[col].astype(str).str.lower() == MISSING_VALUE_STRING)
    
    count_to_impute = is_missing.sum()
    
    if count_to_impute > 0:
        
        # 4. Using .loc[] to replace the missing strings with the new category name
        df.loc[is_missing, col] = NEW_CATEGORY_NAME
        
        print(f"| Imputed '{col}': Replaced {count_to_impute} instances of '{MISSING_VALUE_STRING}' with '{NEW_CATEGORY_NAME}'.")
        imputed_cols += 1
    
print("\n--- Summary ---")
if imputed_cols > 0:
    print(f"✅ Successfully imputed missing values in {imputed_cols} categorical columns.")
else:
    print("ℹ️ No string-based missing values were found.")
```

```python
# Outlier cleaning and capping on numerical columns
print("### Outlier Cleaning (Capping) on Numerical Columns ###")

# Capping logic
def clip_outliers(df, columns):
    for col in columns:
        lower = df[col].quantile(0.01)
        upper = df[col].quantile(0.99)
        df[col] = df[col].clip(lower=lower, upper=upper)
    return df

df = clip_outliers(df, ['Fiyat', 'Net_Metrekare', 'Brüt_Metrekare'])
df.describe()
```

```python
import re

def remove_text1(veri):
    sayilar = [float(s) for s in re.findall(r'\d+', str(veri))]
    
    if len(sayilar) == 2:
        return sum(sayilar) / 2
    
    elif len(sayilar) == 1:
        return sayilar[0]
    
    return None

df_encoded['Binanın_Yaşı'] = df_encoded['Binanın_Yaşı'].apply(remove_text1).astype(float)
```

```python
def remove_text2(data):
    yazi = str(data).lower()
    
    if "giriş" in yazi or "zemin" in yazi or "bahçe" in yazi:
        return 0
    
    sayilar = re.findall(r'\d+', yazi)
    if sayilar:
        return float(sayilar[0])
    return None

df_encoded['Bulunduğu_Kat'] = df_encoded['Bulunduğu_Kat'].apply(remove_text2).astype(float)
```

#### Note on evaluation criteria
Since it's a regression problem, not a classification problem, accuracy and recall are not really meaningful here.  Accuracy only applies to classification. Here, there's no one correct class to get right (e.g sick or not sick), but a whole range of them.  
The target is continous, so we should focus less on getting the exact same number, as there's just too many, and more on getting the least possible error, because at the scale of a house purchase, a model that is off by only a few thousand lira is irrelevant.  
If someone is buying a house and the error is the same price as a pair of shoes or a good jacket, at the scale of the financing the buyer has ready, that's likely to be acceptable. The same goes for someone trying to sell a house: If the model underpriced their house they can just try to negotiate up, and if not, if you're selling a house for a million lira, you probably wouldn't risk the deal for 5k or even 10k.  
So our 3 criteria are as follows:   

&nbsp;&nbsp;&nbsp;&nbsp;Mean Absolute error(MAE): How many lira, plus or minus, is the model off by on average?  
&nbsp;&nbsp;&nbsp;&nbsp;Root Mean Squared Error(RMSE): How big are the mistakes when they are made?  
&nbsp;&nbsp;&nbsp;&nbsp;R^2 score: How much is the model able to guess how different variable affect pricing, where 1 means it understands fully and 0 means it's not much different from making a random guess.

## Hyperparameter Tuning

```python
# Importing functionality
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Defining parameters
param_grid = {
    'n_estimators': [100, 250, 500, 750],    # Number of trees / Boosting stages
    'max_depth': [3, 5, 7, 10],              # Maximum tree depth
    'learning_rate': [0.01, 0.05, 0.1, 0.2], # Step size / Shrinkage factor
    'subsample': [0.6, 0.8, 1.0],            # Subsample ratio of observations/samples
    'colsample_bytree': [0.6, 0.8, 1.0]      # Subsample ratio of columns/features
}
```

```python
# Defining parameters
xgb = XGBRegressor(
    random_state=42,
    n_jobs=-1,
    max_depth=3,          
    n_estimators=300,     
    learning_rate=0.03    
)
```

```python
# Running RandomizedSearchCV
random_search = RandomizedSearchCV(
    estimator=xgb,
    param_distributions=param_grid,
    n_iter=25,
    scoring='neg_mean_squared_error',
    cv=3,
    verbose=2, 
    random_state=42,
    n_jobs=-1
)
```

```python
# Fitting
random_search.fit(x_train, y_train)
```

```python
# Recording best results for later use
best_params = random_search.best_params_
best_score_mse = -random_search.best_score_ 
best_score_rmse = np.sqrt(best_score_mse)
```
